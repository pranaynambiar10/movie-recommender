```python
import ast

def convert_cast(text):
    l=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter<3:
            l.append(i['name'])
        counter+=1

    return l    
```

```python
import ast

def convert_cast(text):
    if isinstance(text, list):  # If it's already a list, return it
        return text
    
    if isinstance(text, float):  # Handle NaN values
        return []
    
    try:
        data = ast.literal_eval(text)  # Convert string to list
        if isinstance(data, list):
            return [i['name'] for i in data if 'name' in i]  # Extract names
        return []
    except (ValueError, SyntaxError):
        return []

# Apply the function
merged_movies['cast'] = merged_movies['cast'].apply(convert_cast)

# Check result
merged_movies.head()
```

```python
merged_movies['cast']=merged_movies['cast'].apply(convert_cast)
merged_movies.head()

```

```python
import ast

def fetch_director(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            l.append(i['name'])
            break

    return l    
```

```python
merged_movies['crew']=merged_movies['crew'].apply(fetch_director)
merged_movies.head()

```

```python
merged_movies['overview']=merged_movies['overview'].apply(lambda x:x.split())
merged_movies.head()
```

```python
import re
import ast

def fix_spacing(text):
    if isinstance(text, list):
        text = " ".join(text)  # Convert list to space-separated string
    if isinstance(text, str):
        return re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return ""

# Apply fix to each column
columns_to_fix = ['overview', 'genres', 'crew', 'keywords', 'cast']
for col in columns_to_fix:
    merged_movies[col] = merged_movies[col].apply(fix_spacing)

# Recreate 'tags' column correctly
merged_movies['tags'] = (
    merged_movies['overview'] + " " +
    merged_movies['genres'] + " " +
    merged_movies['crew'] + " " +
    merged_movies['keywords'] + " " +
    merged_movies['cast']
).apply(fix_spacing)  # Ensure final cleanup

```

```python
merged_movies.head(3)
```

```python
import ast  # To safely convert string representation of lists into Python lists

def extract_keywords(obj):
    if isinstance(obj, str):  # Convert string to a list if needed
        try:
            obj = ast.literal_eval(obj)
        except (ValueError, SyntaxError):
            return ""  # Return empty string if conversion fails
    if isinstance(obj, list):  # Extract "name" values from list of dictionaries
        return " ".join(i["name"] for i in obj if isinstance(i, dict) and "name" in i)
    return ""

# Apply the function to the 'keywords' column
merged_movies['keywords'] = merged_movies['keywords'].apply(extract_keywords)
```

```python
merged_movies['tags'] = (
    merged_movies['overview'].astype(str) + " " +
    merged_movies['genres'].astype(str) + " " +
    merged_movies['crew'].astype(str) + " " +
    merged_movies['keywords'].astype(str) + " " +
    merged_movies['cast'].astype(str)
)
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)    
```

```python
new_df['tags']=new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.replace(" ", ""))  # Remove unnecessary spaces
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(new_df['tags']).toarray()


```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
from thefuzz import process  # ✅ Correct Import

def recommend(movie):
    movie = movie.strip().lower()  # Normalize input
    
    # Step 1: Find the Closest Matching Movie Title
    best_match = process.extractOne(movie, new_df['title'])
    
    if best_match is None or best_match[1] < 80:  # Confidence threshold
        print("Movie not found in dataset!")
        return

    index = new_df[new_df['title'] == best_match[0]].index[0]
    
    # Step 2: Get Similar Movies Based on Cosine Similarity
    distances = sorted(list(enumerate(similarity[index])), key=lambda x: x[1], reverse=True)

    # Step 3: Print Top 5 Recommended Movies
    print(f"Top recommendations for '{best_match[0]}':")
    for i in distances[1:3]:  # Exclude the first (input movie itself)
        print(f"- {new_df.iloc[i[0]].title}")
```

```python
recommend('The Dark Knight Rises')
```
